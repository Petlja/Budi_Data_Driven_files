У ћелијама могу стајати и компликованији кодови од простог рачунања. На пример, ако извршимо код у следећој ћелији, видећемо да се у оквиру Џупитера без проблема покреће и компликованији код, који у себи садржи петље и услове. 

```python
l = [1,2,3,4,5,6] #дефинишемо променљиву l у коју смештамо једну листу
z = [] # правимо празну листу
for i in l: # за сваки елемент листе l
    if i % 2 == 0: #ако је остатак тог елемента при дељењу са 2 једнак 0 (ако је број паран)
        z.append(i) #у празну листу смештамо паран број 
print(z) #штампамо листу парних бројева
```

Ово је важно зато што је "географски" редослед извршавања ћелија секундаран у односу на "хронолошки" редослед њиховох извршавања. Хајде да видимо шта ово практично значи:

Обратите пажњу на неколико претходних ћелија. Прво смо извршили ћелију која је прва одозго и дефинисали смо вредност варијабле **а**. Када смо то једном урадили, **а** ће имати ту вредност током читавог нашег рада, то јест док га не променимо или док не рестартујемо кернел. Исту ту вредност можемо позвати касније, на пример у ћелији испод ње (коју смо извршили осму по реду).

Неважно је да ли је промељива дефинисана у ћелији која тренутно није активна - Џупитер ове вредности памти (као и низ других елемената, са чиме ћете се упознати касније у курсу). На пример, неколико ћелија раније дефинисали смо промељиву l у коју смо сместили листу. Ако смо ћелију покренули, кернел (ускоро ћеш га упознати) и даље памти вредности, тако да можемо поново користити исте промељиве (и из географски удаљених ћелија).  

```python
print(l, z)
```

Због тога је јако важно да имате увид у то шта сте у којој ћелији урадили и када и у томе вам помажу ови бројеви у угластим заградама, који говоре о редоследу извршавања ћелија. Aко ћелија у угластим заградама има неки број, то значи да је успешно извршена, а број представља редни број којим су ћелије извршаване. 

Увек обратите пажњу на редослед извршавања ћелија! Такође, ако видите да је угласта заграда поред ћелије са кодом празна, то значи да се ћелија није извршила. Ако ћелија у себи има звездицу, то значи да се код тренутно извршава.

## Шта је кернел и зашто би неко пожелео да га прекине или рестартује?

Када комуницирамо са Џупитером, то шематски изгледа овако:

![Komunikacija sa Jupyterom](slike/notebook_components.png)

**Кернел** (енгл. "језгро") је "машина" која извршава команде и памти податке. То је главни погон радне свеске.

Свака нова радна свеска коју отворимо:
- отвори нови таб у прегледачу (browser), и
- покрене своју копију кернела.

Према томе, колико радних свески отворимо, толико копија кернела ће бити покренуто.

**Затварањем таба у прегледачу се НЕ затвара кернел који је због те радне свеске отворен!**

Да би се затворио таб **И** кернел, треба одабрати File -> Close and Halt

Рестартовање кернела можемо пожелети да урадимо када пожелимо да из меморије избришемо све учитане податке, вредности додељене варијаблама, и сл.

То радимо тако што у горњем менију кликнемо на *Kernel* и изаберемо опцију *Restart*.

Када рестартујемо Кернел, варијабле које смо до тада дефинисали изгубиће вредности. Сетимо се варијабле а којој смо горе доделили вредност 5. Ако покушаш да покренеш ћелију испод, видећеш да Џупитер и даље памти њену вредност. 

```python
а
```

Међутим, покушај сада да рестартујеш кернел и да поново покренеш ћелију изнад. Видећеш да ће Пајтон пријавити грешку и рећи ће ти да а није дефинисано. 

Поред рестартовања Кернела, често ћемо пожелети и да избришемо све резултате извршавања кода у нашим ћелијама. Видећеш и сам/сама, свеске се јако брзо увећавају и постају све конфузније, па ћеш понекад желети да "кренеш испочетка". Ово можеш урадити тако што ћеш у горњем менију кликнути на дугме Кернел и изабрати опцију Restart & Clear all output. 

Наравно, то што имаш могућност да то урадиш не значи да је то добра пракса па ти саветујемо да се трудиш да лепо организујеш садржај својих свезака, како би другим људима, а и теби, било јасно шта си са одређеним деловима кода и анализама хтео/хела да постигнеш.

Такође, не морате сваки пут рестартовати кернел и покренути избрисати резултат извршења свих ћелија. Можете избрисати и резултат извршења појединачних ћелија тако што ћете кликнути на опцију *Cell->Current Output->Clear*, што ће избрисати резултат извршавања одабране ћелије.

Ако желите да избришете резултат извршавања свих ћелија без рестартовања кернела, што ће избрисати све резултате извршавања, али сачувати вредности варијабли, дефинисане функције и слично, то можете урадити тако што ћете кликнути на  *Cell->All Output->Clear*
