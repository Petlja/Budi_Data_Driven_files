## Џупитер радна свеска

Џупитер је окружење у коме се могу извршавати Пајтон програми. У Џупитеру, твоје анализе организоване су у радне свеске (фајлове са екстензијом .ipynb). Џупитер радна свеска се састоји из низа ћелија, а свака ћелија може да садржи текст, математички израз или низ Пајтон наредби. На тај начин цео процес обраде података је интерактиван, бележимо га у Џупитер радној свесци коју лако можемо поделити са неким и која оставља траг о томе шта је све урађено током анализе одређених података. 

## Интерфејс

Иконоце које се налазе на врху радне свеске ти омогућавају да манипулишеш ћелијама овог окружења на разне начине.

Дугме **Run** користимо за покретање ћелије. Ако је то ћелија са кодом, код ће се извршити, ако је ћелија са текстом, текст ће се исписати. Овај текст који сада читаш написан је у једној ћелији и извршен

У другом пољу са десне стране, бира се да ли ће ћелија бити текстуалног типа (у том случају пише **Markdown**) или ћелија са кодом (**Code**).

Све оно што смо видели у Кратком уводу у Пајтон важи и даље. 

Треба знати да се ово окружење може користити и као најобичнији калкулатор. Рецимо, уколико желиш да одмах израчунаш нешто, можеш додати нову ћелију и у њој извршити неку од основних рачунских операција. Додај ћелију и изврши неку од операција.

У ћелијама могу стајати и компликованији кодови од простор рачунања. На пример, ако извршимо код у следећој ћелији, видећемо да се у оквиру Џупитера без проблема покреће и компликованији код, који у себи садржи петље и услове. 

```python
l = [1,2,3,4,5,6] #дефинишемо променљиву l у коју смештамо једну листу
z = [] # правимо празну листу
for i in l: # за сваки елемент листе l
    if i % 2 == 0: #ако је остатак тог елемента при дељењу са 2 једнак 0 (ако је број паран)
        z.append(i) #у празну листу смештамо паран број 
print(z) #штампамо листу парних бројева
```

## Шта је кернел и зашто би неко пожелео да га прекине или рестартује?

Када комуницирамо са Џупитером, то шематски изгледа овако:

![Komunikacija sa Jupyterom](slike/notebook_components.png)

**Кернел** (енгл. "језгро") је "машина" која извршава команде и памти податке. То је главни погон радне свеске.

Свака нова радна свеска коју отворимо:
- отвори нови таб у прегледачу (browser), и
- покрене своју копију кернела.

**Затварањем таба у прегледачу се НЕ затвара кернел који је због те радне свеске отворен!**
